```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/User/OneDrive/Desktop/DA Intern @ ApexPlanet/Task-1/raw_dataset.csv")

df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()

# Remove duplicates
df = df.drop_duplicates()
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], errors='coerce')
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
```

```python
df['Customer_Name'] = df['Customer_Name'].str.title()
df['Product'] = df['Product'].str.title()
df['City'] = df['City'].str.title()
df['Category'] = df['Category'].str.title()
df['Payment_Method'] = df['Payment_Method'].str.upper()
```

```python
df['Customer_Age'] = datetime.now().year - df['DOB'].dt.year
```

```python
df['Total_Revenue'] = df['Quantity'] * df['Unit_Price']
```

```python
df['Month'] = df['Transaction_Date'].dt.month_name()
```

```python
df.to_csv("C:/Users/User/OneDrive/Desktop/DA Intern @ ApexPlanet/Task-1/cleaned_dataset.csv", index=False)
```
